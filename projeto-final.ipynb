```python
# - vizu 2  - Quantidade de Estabelecimentos com Berçário/creche

colunas_bercario_creche = ['Âmbito', 'UF',
                           '2.2 Possui berçário e/ou centro de referência materno-infantil? (apenas para estabelecimentos com vagas para mulheres)',
                           '2.2 Possui berçário e/ou centro de referência materno-infantil? (apenas para estabelecimentos com vagas para mulheres) | Sim. Capacidade de bebês: | Quantidade',
                           '2.3 Possui creche? (apenas para estabelecimentos com vagas para mulheres)',
                           '2.3 Possui creche? (apenas para estabelecimentos com vagas para mulheres) | Sim. Capacidade de crianças: | Quantidade']

# Criando um novo DataFrame com as colunas de berçário e creche
df_bercario_creche = df_original[colunas_bercario_creche].copy()

# Renomeando as colunas para facilitar o acesso
df_bercario_creche.columns = ["Ambito", "UF", "Possui_Bercario", "Capacidade_Bercario", "Possui_Creche", "Capacidade_Creche"]

# Somando as colunas relevantes para obter a quantidade de estabelecimentos com berçário e creche
total_bercario = df_bercario_creche['Possui_Bercario'].eq('Sim. Capacidade de bebês:').sum()
total_creche = df_bercario_creche['Possui_Creche'].eq('Sim. Capacidade de crianças:').sum()
total_bercario_Nao = df_bercario_creche['Possui_Bercario'].eq('Não').sum()
total_creche_Nao = df_bercario_creche['Possui_Creche'].eq('Não').sum()

# Somando os totais para obter a quantidade total de estabelecimentos com berçário e creche
total_estabelecimentos_bercario_creche = total_bercario + total_creche
total_estabelecimentos_bercario_creche_Nao = total_bercario_Nao + total_creche_Nao


# Exibindo o resultado
print(f"Quantidade de Estabelecimentos com Berçário: {total_bercario}")
print(f"Quantidade de Estabelecimentos com Creche: {total_creche}")
print(f"Total de Estabelecimentos com Berçário e/ou Creche: {total_estabelecimentos_bercario_creche}")
print(f"Quantidade de Estabelecimentos que não possui Berçário: {total_bercario_Nao}")
print(f"Quantidade de Estabelecimentos que não possui Creche: {total_creche_Nao}")
print(f"Total de Estabelecimentos que não possuem Berçário e/ou Creche: {total_estabelecimentos_bercario_creche_Nao}")


```
