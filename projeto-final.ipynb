```python
#vizu média de idade ponderada para cada faixa etária e Calculando a média de idade ponderada pela quantidade de pessoas media_idade = (
# Substitua 'SEU_CAMINHO/SEU_ARQUIVO.csv' pelo caminho real do seu arquivo CSV ou use o método de leitura apropriado para o seu caso
# df_sistema_prisional_feminino_brasil = pd.read_csv('SEU_CAMINHO/SEU_ARQUIVO.csv')

# Seleção das colunas relevantes
colunas_idade_tempo_pena = [
    "UF",
    "5.1 Quantidade de pessoas privadas de liberdade por faixa etária | 18 a 24 anos | Feminino",
    "5.1 Quantidade de pessoas privadas de liberdade por faixa etária | 25 a 29 anos | Feminino",
    "5.1 Quantidade de pessoas privadas de liberdade por faixa etária | 30 a 34 anos | Feminino",
    "5.1 Quantidade de pessoas privadas de liberdade por faixa etária | 35 a 45 anos | Feminino",
    "5.12 Quantidade de pessoas privadas de liberdade por tempo total de penas (presos/as condenados/as e ) | Mais de 2 até 4 anos (inclusive) | Feminino",
    "5.12 Quantidade de pessoas privadas de liberdade por tempo total de penas (presos/as condenados/as e ) | Mais de 4 até 8 anos (inclusive) | Feminino",
    "5.12 Quantidade de pessoas privadas de liberdade por tempo total de penas (presos/as condenados/as e ) | Mais de 8 até 15 anos (inclusive) | Feminino",
    "5.12 Quantidade de pessoas privadas de liberdade por tempo total de penas (presos/as condenados/as e ) | Mais de 15 até 20 anos (inclusive) | Feminino",
    "5.12 Quantidade de pessoas privadas de liberdade por tempo total de penas (presos/as condenados/as e ) | Mais de 20 até 30 anos (inclusive) | Feminino",
]

# Criando um DataFrame com as colunas selecionadas
df_analise_idade_tempo_pena = df_sistema_prisional_feminino_brasil[colunas_idade_tempo_pena]

# Definindo os índices numéricos das colunas
faixa_etaria_indices = [1, 2, 3, 4]  # Substitua pelos índices reais das faixas etárias
pena_indices = [5, 6, 7, 8, 9]  # Substitua pelos índices reais das penas

# Calculando a média de idade ponderada para cada faixa etária e pena
for i in range(len(faixa_etaria_indices)):
    coluna_qtde = colunas_idade_tempo_pena[faixa_etaria_indices[i]]
    coluna_tempo_pena = colunas_idade_tempo_pena[pena_indices[i]]
    
    # Calculando a média de idade ponderada pela quantidade de pessoas
    media_idade = (
        (df_analise_idade_tempo_pena[coluna_qtde] * df_analise_idade_tempo_pena.apply(lambda row: int(row[coluna_qtde].split()[0]) if isinstance(row[coluna_qtde], str) else row[coluna_qtde], axis=1)).sum() +
        (df_analise_idade_tempo_pena[coluna_tempo_pena] * df_analise_idade_tempo_pena.apply(lambda row: (int(row[coluna_tempo_pena].split()[3]) + int(row[coluna_tempo_pena].split()[6])) / 2 if isinstance(row[coluna_tempo_pena], str) else row[coluna_tempo_pena], axis=1)).sum()
    ) / (df_analise_idade_tempo_pena[coluna_qtde].sum() + df_analise_idade_tempo_pena[coluna_tempo_pena].sum())

    # Simplificando o print
    faixa_etaria = coluna_qtde.split("|")[-2]
    pena = coluna_tempo_pena.split("|")[-2]
    print(f"Média de idade para a faixa etária {faixa_etaria} e pena {pena} =  {media_idade:.2f} anos")

```
